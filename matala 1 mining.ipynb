```python
import requests
from bs4 import BeautifulSoup

def get_cars(keywords):
    pages_list = []
    cars_list = []
    base_url = 'https://www.ad.co.il/car?keyword=' + keywords

    try:
        # Fetch the first page to find pagination links
        response = requests.get(base_url)
        if response.status_code != 200:
            print("Error:", response.status_code)
            return None

        results_page = BeautifulSoup(response.content, 'html.parser')
        pages = results_page.find_all('li', {'class': 'page-item mx-1 d-none d-sm-block'})
        
        for page in pages:
            page_link = "https://www.ad.co.il" + page.find('a').get('href')
            pages_list.append(page_link)

        pages_list.insert(0, base_url)

        for page_url in pages_list:
            response = requests.get(page_url)
            if response.status_code != 200:
                print("Error:", response.status_code)
                continue

            results_page = BeautifulSoup(response.content, 'html.parser')
            cars = results_page.find_all('div', {'class': 'card-body p-md-3'})
            
            for car in cars:
                car_link = "https://www.ad.co.il" + car.find('a').get('href')
                car_name = car.find('a').get_text(strip=True)
                try:
                    car_price = car.find('div', {'class': 'price ms-1'}).get_text(strip=True)
                except:
                    car_price = ''
                cars_list.append((car_name, car_link, car_price))
        
        return cars_list

    except Exception as e:
        print("Error:", e)
        return None
```

```python
def get_car_info(car_link):
    car_dict = dict()
    import requests
    from bs4 import BeautifulSoup
    try:
        response = requests.get(car_link)
        if response.status_code != 200:
            return car_dict
        result_page = BeautifulSoup(response.content,'html.parser')
        car_list = []
        table_rows = result_page.find_all('tr')
        for row in table_rows:
            cells = row.find_all('td')
            row_data = []
            for cell in cells:
                row_data.append(cell.get_text(strip=True))
            car_list.append(row_data)
        for item in car_list:
            key = item[0]
            value = item[1]
            car_dict[key] = value
        pics_list = []
        pics = result_page.find_all('div', {'class': 'justify-content-center px-1'})
        for pic in pics:
            pic_num = pic.find('jpg', {'class': 'desktop-thumbnail bg-video'})
            pics_list.append(pic_num)
        counter = len(pics_list)
        car_dict['num_pics'] = counter
        dates_list = []
        dates = result_page.find_all('div', {'class': 'px-3'})
        for date in dates:
            dates_list.append((date.get_text(strip=True)))
        car_dict['Cre_date'] = dates_list[0]
        car_dict['Repub_date'] = dates_list[1]
        exp_list = []
        try:
            exp = result_page.find_all('div', {'class': 'p-3'})
            for e in exp:
                p_tag = e.find('p', {'class': 'text-word-break'})
                if p_tag:
                    expl = p_tag.get_text(strip=True)
                    exp_list.append(expl)
            if exp_list:
                car_dict['Description'] = exp_list[0]
            else:
                car_dict['Description'] = ''
        except Exception as e:
            car_dict['Description'] = str(e)
        return car_dict
    except:
        return None
```

```python
def get_all_cars(keywords):
    results_list = []  
    all_cars = get_cars(keywords)
    for car in all_cars:
        car_dict = {}
        car_dict['name'] = car[0]
        car_dict['price'] = car[2]
        info = get_car_info(car[1])
        for key, value in info.items():
            car_dict[key] = value
        results_list.append(car_dict)
    return results_list
```
